```python
unique_years = sorted(df["released_year"].unique())

mean_dance = [df.loc[df["released_year"] == year, "danceability_%"].mean() for year in unique_years]
mean_val = [df.loc[df["released_year"] == year, "valence_%"].mean() for year in unique_years]
mean_ener = [df.loc[df["released_year"] == year, "energy_%"].mean() for year in unique_years]
mean_acoustic = [df.loc[df["released_year"] == year, "acousticness_%"].mean() for year in unique_years]
mean_instrument = [df.loc[df["released_year"] == year, "instrumentalness_%"].mean() for year in unique_years]
mean_live = [df.loc[df["released_year"] == year, "liveness_%"].mean() for year in unique_years]
mean_speech = [df.loc[df["released_year"] == year, "speechiness_%"].mean() for year in unique_years]

df_means = pd.DataFrame({"Years": unique_years, "danceability_%": mean_dance, "valence_%": mean_val,
                         "energy_%": mean_ener, "acousticness_%": mean_acoustic, "instrumentalness_%": mean_instrument,
                         "liveness_%": mean_live, "speechiness_%": mean_speech})
df_means
```

```python
import seaborn as sns


plt.figure(figsize=(14, 8))
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 10))

axes[0][0].plot(df_means["Years"], df_means["danceability_%"], linewidth=2)
axes[0][0].set_title('Danceablity')
axes[0][0].set_xlabel('Years')
axes[0][0].set_ylabel('Percentage (%)')


axes[0][1].plot(df_means["Years"], df_means["valence_%"], linewidth=2)
axes[0][1].set_title('Valence')
axes[0][1].set_xlabel('Years')
axes[0][1].set_ylabel('Percentage (%)')


axes[0][2].plot(df_means["Years"], df_means["energy_%"], label='Energy', linewidth=2)
axes[0][2].set_title('Energy')
axes[0][2].set_xlabel('Years')
axes[0][2].set_ylabel('Percentage (%)')

axes[1][0].plot(df_means["Years"], df_means["acousticness_%"], label='Acousticness', linewidth=2)
axes[1][0].set_title('Acousticness')
axes[1][0].set_xlabel('Years')
axes[1][0].set_ylabel('Percentage (%)')

axes[1][1].plot(df_means["Years"], df_means["instrumentalness_%"], label='Instrumentalness', linewidth=2)
axes[1][1].set_title('Instrumentalness')
axes[1][1].set_xlabel('Years')
axes[1][1].set_ylabel('Percentage (%)')

axes[1][2].plot(df_means["Years"], df_means["liveness_%"], label='Liveness', linewidth=2)
axes[1][2].set_title('Liveness')
axes[1][2].set_xlabel('Years')
axes[1][2].set_ylabel('Percentage (%)')


axes[2][0].plot(df_means["Years"], df_means["speechiness_%"], label='Speechiness', linewidth=2)
axes[2][0].set_title('Speechiness')
axes[2][0].set_xlabel('Years')
axes[2][0].set_ylabel('Percentage (%)')


plt.delaxes(axes[2][1])
plt.delaxes(axes[2][2])
plt.tight_layout()



plt.show()
```
